```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('dataset.csv')
print(df.shape)
df.head(5)
```

```python
for col in df:
    print(df[col].value_counts())
```

```python
df.isnull().sum()
```

```python
df_sym=pd.read_csv('Symptom-severity.csv')
df_sym.head(5)
```

```python
rank=df_sym['Symptom'].unique()
```

```python
one_d=df[df.columns].values.flatten()
```

```python
df_F=pd.Series(one_d)
df_F= df_F.str.strip()
df_F=df_F.values.reshape(df.shape)

df_N=pd.DataFrame(df_F, columns=df.columns)
value=df_N.values
df_N.head(100)
```

```python
for r in range(len(rank)):
    value[value == rank[r]] = df_sym[df_sym['Symptom'] == rank[r]]['weight'].values[0]
    
df_R=pd.DataFrame(value, columns=df.columns)
df_R.head()
```

```python
for col in df_R:
    print(df_R[col].value_counts())
```

```python
df_R.replace(dict.fromkeys(['foul_smell_of urine','dischromic _patches','spotting_ urination'],0),inplace=True)
```

```python
df_R.fillna(0,inplace=True)
df_R.head()
```

```python
X=df_R.iloc[:,1:]
y=df_R.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
model=DecisionTreeClassifier(random_state=42)
model=model.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(18.5,16))
plot_tree(model, filled=True, rounded=True,
        class_names = True,
         feature_names=X_train.columns);
```

```python
fig.savefig("decistion_tree.pdf")
```

```python
y_pre=model.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pre))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
cm=confusion_matrix(y_test,y_pre)
plt.figure(figsize = (18,12))
ax = sns.heatmap(cm, annot=True, fmt='g');
ax.set_title('Confusion Matrix');
ax.set_xlabel('Predicted Disease')
ax.set_ylabel('Actual Disease');
```
